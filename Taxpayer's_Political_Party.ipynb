```
# getting all the categorical columns
cat_cols = tax_data.select_dtypes(include = 'object').columns  
cat_cols 
```

```
for col in ['Married', 'CollegGrads', 'Cars']:
    tax_data[col] = tax_data[col].astype('category')
```

```
tax_data.dtypes
```

```
# getting all the categorical columns
cat_cols = tax_data.select_dtypes(include = 'category').columns 
cat_cols.size
```

```

# earlier we have collected all the categorical columns in cat_cols
tax_data.shape
tax_data[cat_cols] = tax_data[cat_cols].fillna(tax_data[cat_cols].mode().iloc[0])
tax_data = pd.get_dummies(tax_data, columns=cat_cols)
tax_data.shape
tax_data.head()
```

```
tax_data.head(10)
```

```
tax_data['PoliticalParty'] = tax_data['PoliticalParty'].astype(str)
```

```
tax_data.head(10)
```

```
# Converting gender colum - 'Male' to 1 and 'Female' 0
tax_data['PoliticalParty']=tax_data['PoliticalParty'].replace({"Democrat":0,"Republican":1, "Independent":2})
```

```
tax_data.head(10)
```

```
X_train_r = tax_data.copy()
X_train_r.head()

```

```
del X_train_r['PoliticalParty']
X_train_r.head()
```

```
y_train_r = tax_data['PoliticalParty']
y_train_r.head(10)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train_r, y_train_r)
```
